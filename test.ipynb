```python
from src import FFNN, layer, activation_functions, initializers, loss_functions, optimizers

activation_function = activation_functions.ReLU()  # Example activation function
weight_initializer = initializers.ConstantInitializer(value=1)  # Example initializer
bias_initializer = initializers.ConstantInitializer(value=0)  # Example bias initializer
loss_function = loss_functions.MSELoss()  # Example loss function
optimizer = optimizers.SGD(learning_rate=0.01)

layers = [layer.Layer(input_size=3, output_size=2, activation_function=activation_function,
                weight_initializer=weight_initializer, bias_initializer=bias_initializer)]

model = FFNN.FFNN(layers=layers, loss_function=loss_function, optimizer=optimizer)

```

```python
import numpy as np

x = np.array([[0.1, 0.2, 0.3], [0.2, 0.4, 0.6]])

y = np.array([[0.3, 0.6], [0.6, 1.2]])

for _ in range(100):
    model.train(x, y)



predictions = model.forward_pass(x)
print("Predictions:\n", predictions)
```
